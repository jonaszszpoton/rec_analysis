### 2. Monte Carlo simulation - positive class alternative outcome evaluation:
Assesing ground truth probability of recidivism on non-incarcerated inamates only. Then applying that base probability to all inmates. Checking the difference in predictions. The goal is to check, in what percentage of cases an inmate with high model score (full dataset model training or RisCanvi score) would have recidiviated if not for the incarceration (estimating injustice score). The goal is to reliably estimate positive class error without 'feedback loop bias' (real recidivism numbers among higher score inmates tend to be lower - if incarcerated, cannot recidiviate).
